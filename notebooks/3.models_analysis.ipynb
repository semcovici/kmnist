## Create kernel images

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def plot_kernels(model, layer, file_name):

    # Extraindo os pesos da camada convolucional
    filters, biases = model.layers[layer].get_weights()
    n_filters = filters.shape[-1]

    # Normalizando os filtros para a faixa [0, 1]
    filters_min = filters.min()
    filters_max = filters.max()
    filters = (filters - filters_min) / (filters_max - filters_min)

    # Definindo nmero de colunas e linhas
    cols = 8  # Definindo um nmero fixo de colunas para a grid
    rows = math.ceil(n_filters / cols)  # Calculando o nmero de linhas

    # Ajustando o tamanho da figura com base no nmero de filtros
    fig, axs = plt.subplots(rows, cols, figsize=(cols, rows))
    axs = axs.ravel()  # Converter a grade de eixos para uma lista plana
    
    for i in range(n_filters):
        f = filters[:, :, :, i]
        axs[i].imshow(f[:, :, 0], cmap='gray')
        axs[i].axis('off')

    # Remover subplots extras que no tm filtros associados
    for i in range(n_filters, rows * cols):
        axs[i].axis('off')

    plt.subplots_adjust(wspace=0.1, hspace=0.1)  # Ajusta o espaamento entre os subplots
    plt.savefig(IMAGES_PATH + f'kernels_plt/{file_name}', bbox_inches='tight')
    plt.show()

    plt.clf()


plot_kernels(model_cnn, layer=0, file_name='model_cnn_layer_0_kernel_plot.png')
plot_kernels(model_cnn, layer=1, file_name='model_cnn_layer_1_kernel_plot.png')


plot_kernels(model_cnn_wo_poll, layer=0, file_name='model_cnn_without_polling_layer_0_kernel_plot.png')
plot_kernels(model_cnn_wo_poll, layer=1, file_name='model_cnn_without_polling_layer_1_kernel_plot.png')

plot_kernels(model_cnn_aug, layer=0, file_name='model_cnn_aug_layer_0_kernel_plot.png')
plot_kernels(model_cnn_aug, layer=1, file_name='model_cnn_aug_layer_1_kernel_plot.png')
```
