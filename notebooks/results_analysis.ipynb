```python
import pandas as pd
import json
import matplotlib.pyplot as plt

from sklearn.metrics import classification_report
```

```python
RESULTS_PATH = "../results/"
```

```python
with open(RESULTS_PATH + "history/fcnn_history.json", 'r') as file:
    
    fcnn_history = json.loads(file.read())
    
with open(RESULTS_PATH + "history/cnn_history.json", 'r') as file:
    
    cnn_history = json.loads(file.read())
    
with open(RESULTS_PATH + "history/benchmark_cnn_history.json", 'r') as file:
    
    benchmark_cnn_history = json.loads(file.read())
```

```python
test_results_benchmark = pd.read_csv(RESULTS_PATH + 'test/cnn_benchmark_test_results.csv')
train_results_benchmark = pd.read_csv(RESULTS_PATH + 'train/cnn_benchmark_train_results.csv')

train_results_cnn = pd.read_csv(RESULTS_PATH + 'train/cnn_train_results.csv')
test_results_cnn = pd.read_csv(RESULTS_PATH + 'test/cnn_test_results.csv')

train_results_fcnn = pd.read_csv(RESULTS_PATH + 'train/fcnn_train_results.csv')
test_results_fcnn = pd.read_csv(RESULTS_PATH + 'test/fcnn_test_results.csv')
```

## ANALYSIS

### BENCHMARK CNN

```python
# summarize history for accuracy
plt.plot(benchmark_cnn_history['accuracy'])
plt.plot(benchmark_cnn_history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()


# summarize history for loss
plt.plot(benchmark_cnn_history['loss'])
plt.plot(benchmark_cnn_history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
print('################################# TRAIN RESULTS #################################')
print(classification_report(train_results_benchmark['y_train'], train_results_benchmark['y_train_pred']))

print('################################# TEST RESULTS #################################')
print(classification_report(test_results_benchmark['y_test'], test_results_benchmark['y_test_pred']))
```

### FCNN

```python
# summarize history for accuracy
plt.plot(fcnn_history['accuracy'])
plt.plot(fcnn_history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()


# summarize history for loss
plt.plot(fcnn_history['loss'])
plt.plot(fcnn_history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(fcnn_history['learning_rate'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
print('################################# TRAIN RESULTS #################################')
print(classification_report(train_results_fcnn['y_train'], train_results_fcnn['y_train_pred']))

print('################################# TEST RESULTS #################################')
print(classification_report(test_results_fcnn['y_test'], test_results_fcnn['y_test_pred']))
```
